# Loading Libraries

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

# Performing CV

### Reading Image

```python
image = mpimg.imread('../input/animals/cat.jpg')
plt.imshow(image)
```

### Grey Image

To decrease the complexity of the image and make the handling easier we will be converting the image to a grayscale image with the help of the `cvtColor` function.

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
```

### Inverted Grey Image

Now the next step is to invert the image. Because when an image is inverted it helps in studying the image with more precision and detail.  
The more detailed study is important for the sketch generation as the sketches need to be accurate and good.

```python
inverted_image = 255 - gray_image
plt.imshow(inverted_image, cmap='Greys_r')
```

### Blurred Image

In addition to this, we will also be smoothing the image to make sure the sketch we obtain is less edgy and smooth.

```python
inverted_blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('inverted gray')
ax1.imshow(inverted_image, cmap='Greys_r')

ax2.set_title('inverted blurred image')
ax2.imshow(inverted_blurred, cmap='Greys_r')
```

### Pencil Sketch

Now that the whole image processing is done, we will be passing the previous outputs to the function and pass the right and accurate parameters to make the appropriate changes to the image.

```python
final = cv2.divide(gray_image, 255 - inverted_blurred, scale=240)
plt.figure(figsize=(15,15))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```

### All images in one frame

```python
plt.figure(figsize=(30,30))
plt.subplot(1,5,1)
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,5,2)
plt.imshow(gray_image, cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,5,3)
plt.imshow(inverted_image,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.subplot(1,5,4)
plt.imshow(inverted_blurred,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.subplot(1,5,5)
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
